```python
# 2.1) Histogram
# Tek bir verinin nasıl dağıldığını görmek için kullanabilecek çizim yöntemi histogram grafiğidir.
```

```python
# Histogram mantığı;
import matplotlib.pyplot as plt
hist = plt.imread("histogram.png")
plt.imshow(hist)
plt.axis("off")
plt.show()
```

```python
# Data setindeki değerleri bir sayı doğrusunda gösterdiğimi varsay.
# Aralıklar değiştikçe yığılım değişecektir.
# Belirlendğin her aralığa yüksekliği o aralıktaki gözlem sayısına eşit birer sütun
# çizersem veri kümesinin histogramını elde etmiş olurum.
# Bunun için matplotlib kütüphanesinden .hist() metodu kullanılır.
# bins argümanı ile kaç aralık istediğimizi belirleriz.
import matplotlib.pyplot as plt
x= [0.55, 4.22, 3.29, 3.4, 2.9, 3.2, 4.3, 8.8, 5.1, 9.6,
   3.8, 5, 4.2, 0.43, 2.96, 4.7, 8.9, 5.4, 8.1, 5.27]
plt.hist(x, bins=10)
plt.show()
```

```python
plt.hist(x, bins=5)
plt.text(1, 7, "Histogram Grafiği")
plt.show()
```

```python
# .plot metodu ile de histogram grafiği çizdirebilirim.
import seaborn as sns  
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris") 
plt.hist(iris["petal_length"], bins=20)
plt.xlabel("Taç Yaprak Boyu")
plt.ylabel("Frekans")
plt.show()
```

```python
iris.hist(column="petal_length")
```

```python
plt.hist(iris["petal_length"], density=True) # Kutudaki veri sayısı / toplam veri sayısı
plt.xlabel("Taç Yaprak Boyu (cm)")
plt.ylabel("Frekans")
plt.show()
# Tüm gözlemlerin yaklaşık yüzde 40'ında Taç Yaprak boyunun 1- 1.5 cm aralığında olduğunu görüyoruz.
```

```python
iris.plot.hist(alpha=0.7)  # alpha ile opaklığı ayarladım. (Renk kons.)
plt.show()
```

```python
iris.hist()
plt.show()
```

```python
# 2 değişkenden oluşan verilerin histogramlarını çizdirebiliriz. plt.hist2d()
# Bu grafikler sıcaklık haritası olarakta isimlendirilir.
import matplotlib.pyplot as plt
from sklearn import datasets
iris = datasets.load_iris()

X = iris.data[:, 2]
Y = iris.data[:, 3]

plt.subplot(1, 2, 1)
plt.scatter(X,Y)
plt.xlabel = ("petal_length(cm)")
plt.ylabel = ("petal_width(cm)")
plt.title("Nokta Grafik")
plt.subplot(1, 2, 2)
plt.hist2d(X, Y, cmap="plasma")
plt.colorbar()
plt.title("İki Boyutlu Histogram")
plt.show()
```

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

boston = datasets.load_boston()
bosdf= pd.DataFrame(boston.data)
bosdf.columns = boston.feature_names
bosdf.columns
```

```python
bosdf["CRIM"] = boston.target # ?
sns.jointplot(x="LSTAT", y="CRIM", data=bosdf)
plt.show()
```

```python
sns.jointplot(x="LSTAT", y="CRIM", data=bosdf, kind="reg")
plt.show()
```

```python
# Seaborn çizim temasını kullanmak için;
sns.set()
sns.jointplot(x="LSTAT", y="CRIM", data=bosdf, kind="reg")
plt.show()
```

```python
# Kutu grafiği çizdirmenin bir başka yöntemi;
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

iris  = sns.load_dataset("iris")
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.show()
```

```python
# Kutu grafiği üzerine verileri çizdirmekte mümkün.
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

iris  = sns.load_dataset("iris")
sns.boxplot(x="species", y="sepal_length", data=iris)
sns.swarmplot(x="species", y="sepal_length", data=iris, color="k", alpha=0.7)
plt.show()
```

```python
# 2.2) Keman (Violin) Grafiği
sns.violinplot(x="species", y="sepal_length", data=iris)
plt.show()
```

```python
# inner = None ---->  Kutu grafiğini kaldırır.
# .swarmplot yerinde stripplot
# jitter ------>  Verilerin tek bir doğruda değilde sağa sola yayılarak çıktısı alınır
sns.violinplot(x="species", y="sepal_length", data=iris, inner=None)
sns.stripplot(x="species", y="sepal_length", data=iris, jitter=True)
plt.show()
# Rengi düzenlemek gerekti;
```

```python
sns.violinplot(x="species", y="sepal_length", data=iris, inner=None)
sns.stripplot(x="species", y="sepal_length", data=iris, jitter=True, color="black")
plt.show()
```
