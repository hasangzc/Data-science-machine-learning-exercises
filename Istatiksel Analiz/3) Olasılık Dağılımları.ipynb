```python
#  3.1) Bernoulli Dağılımı
# En basit dağılım olan Bernoulli dağılımı sonucun sadece 2 farklı değerden birisi olduğu rassal değişkenleri
# ifade eder.
# Genelleştirip sonucu 1 ya da 0 olarak düşünelim.
# Sonucun 1 olması olasılığı p ise 0 olma olasılığı 1-p' dir.
```

```python
#  3.2) Binom Dağılımı
# Binom dağılımını çok sayıda Bernoulli denemesi olarak düşünebiliriz.
# Binom dağılımında sonuç değer çok sayıda Bernoulli denemesinden elde edilen başarıların toplamadır.
```

```python
# Diyelimki elimizde hileli bir para var. Yazı gelme olasılığı = 0.6
# zarı 3 defa atalım                      Tura gelme olasılığı = 0.4

#  p(x=0) = komb(3 0)*(0.4^0)*(0.6^3) = 0.216
#  p(x=1) = komb(3 1)*(0.4^1)*(0.6^2) = 0.432
#  p(x=2) = komb(3 2)*(0.4^2)*(0.6^1)  = 0.288
#  p(x=3) = komb(3 3)*(0.4^3)*(0.6^0) = 0.064


# 0.216 + 0.432 + 0.288 + 0.064 = 1
# Yukarıdaki olasılık dağılımları aynı zamanda olasılık kütle fonksiyonları olarak adlandırılır.
```

```python
# Bulduklarımızı genelleştirirsek Binom Dağılımı;
# p(X=k) = komb(n k) * (p^k) * ((1-p)^(n-k))

# n = Toplam Bernoulli denemesi     # k = adet başarı elde etme olasılığı
# p = Her bir Bernoulli denemesinde başarı olasılığı
# 1 -p = Her bir Bernoulli denemesinde başarısızlık olasılığı
```

```python
# Şimdi Pythonda bir binom deneyi gerçekleştirelim;
# 3 adet madeni para atma(hileli) deneyi; Yazı = 0 Tura= 1 ile gösterilsin;
import numpy as np
paralar = np.random.random(size=3)
paralar
paralar > 0.6
```

```python
# Yüzde 66' si tura geldi. Halbuki yukarıdaki hesapta 3 paranın 2 sinin tura gelme olasılığını
# yaklaşık 28.8 bulmuştuk.
# Şimdi aynı deneyi birçok defa tekrar edelim;
iki_tura_sayisi = 0
for i in range(0, 1000000):
    tura = np.random.random(size=3) > 0.6
    if np.sum(tura) == 2:
        iki_tura_sayisi +=1
iki_tura_sayisi / 1000000
# Aynı deneyi  milyon defa tekrarladığımda 2 adet tura gelme oranı yüzde 28.78 oldu.
```

```python
# Pythonda Binom dağılımını simüle etmek için np.random.binomial() fonksiyonu kullanılır.
# Örnek verecek olursak;
# Hilesiz bir madeni paranın 5 defa atıldığını düşünelim. Binom deneyinide 10 defa tekrar edelim.
np.random.binomial(n=5, p=0.5, size=10)
```

```python
# Şimdi de az önce for döngüsü ile gerçekleştirdiğimiz deneyi np.random.binomial() ile yapalım.
binom_deney = np.random.binomial(n=3, p=0.6, size=1000000)
binom_deney
```

```python
# Şimdi elde ettiğimiz sonucun histogramını çizdirelim.
import matplotlib.pyplot as plt
plt.hist(binom_deney)
plt.show()
```

```python
# Şimdi atılan paranın tura gelme olasılığı 0.1 olsun ve bir defada 100 para atalım.
# Bu deneyi 1000000 defa tekrarlayalım.
binom = np.random.binomial(n=100, p=0.1, size=1000000)
plt.hist(binom, edgecolor="k", align="mid", bins=30)
plt.show()
```

```python
#  3.3) Geometrik Dağılım
# N adet Bernoulli denemesinden ilk başarıyı elde edene kadar geçecek başarısız deneme sayısı 
# geometrik dağılım olarak hesaplanır.
# Bir madeni para atılmasında başarı tura gelmesi olsun. İlk defa tura gelene kadar karşılaşılan yazı
# sayısı geometrik dağılımdır.
```

```python
# p(X=k) = ((1-p)^(k-1))*p
# p: başarı olasılığı   k: ilk başarıya kadar ki toplam deneme sayısı
import numpy as np
import matplotlib.pyplot as plt

geom = np.random.geometric(p=0.05, size=1000000)
plt.hist(geom, bins=50)
plt.title("Geometrik Dağılım")
plt.show()
```

```python
#  3.4) Poisson Dağılımı
# Bir olayın belirli bir süre içerisinde herhangi bir sayıda gerçekleşme olasılığını veren ancak 
# olayın gerçekleşme zamanının önceki olayın gerçekleşme zamanından tamamen bağımsız olduğu dağılımlara
# Poisson dağılımı denir.
```

```python
# Günlük hayatta karşılaştığımız birçok ayrık olay Poisson dağılımı ile modellenebilir.
# Bir internet sayfasına giren kullanıcı sayısı, gün içerisinde aldığımız e-postalar, bir şehirde 
# gerçekleşen doğumlar, Dünya'ya düşen meteorlar, bir bölgede gerçekleşen trafik kazaları........ 
```

```python
# Poisson Dağılımında bir değişken vardır: incelenen olayın belirli bir süredeki gerçekleşme sayısı
# Örneğin bir işyerine 1 saatte giren müşteri sayısı gibi.
# Bu değişken genellikle λ(Lambda) ile gösterilir.
```

```python
# Varsayalımki işyerimize saatte ortalama 10 müşteri giriyor olsun.
import matplotlib.pyplot as plt
import numpy as np
pois = np.random.poisson(lam=10, size=1000000)
plt.hist(pois, edgecolor="k", align="mid", bins=30, density=True) # density(yoğunluk) yüzdelik aldık sonucu
plt.margins(0.01) # Kenar Boşlukları
plt.show()
```

```python
# Ortalaması 0 ve standart sapması 1 olan bir normal dağılımı inceleyelim.
# Bu dağılıma Standart normal dağılım denir.
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

mu = 0
sigma = 1

x = np.linspace(mu- 3*sigma, mu + 3*sigma, 100)
y = norm.pdf(x, mu, sigma)  # Bu metodla -3 ve +3 arasındaki 100 değerin yukarıdakı eşitliğe göre
#                               normal dağılım olasılık yoğunluk değerlerini hesaplattık.

plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Olasılık Yoğunluk  Fonksiyonu")
plt.margins(0.0)
plt.title("Normal Dağılım Standart Sapma Aralıkları")
plt.show()
```
