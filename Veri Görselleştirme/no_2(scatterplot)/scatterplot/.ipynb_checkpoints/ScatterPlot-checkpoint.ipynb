```python
import matplotlib.pyplot as plt
import numpy as np

yillar = np.arange(2002, 2020)
buyume = [6.4, 5.6, 9.6, 9.0, 7.1, 5.0, 0.8, -4.7, 8.5, 11.1, 4.8, 8.5, 5.2, 6.1, 3.2, 7.5, 2.8, 0.9]

plt.scatter(yillar, buyume)
plt.xticks([2002, 2004, 2006, 2008, 2010, 2012, 2014, 2016, 2018])
plt.yticks([-5.0, 0.0, 5.0, 10.0], ["-%5", "%0", "%5", "%10"])
# yticks: 1.arguman; Eksen üzerindeki değerlerin hangi noktalarda olacağı
#         2.arguman; Bu noktalarda ne yazacağı !  Eğer bu arguamanı girmezsen direk sayı değerleri işlenir yazıya.
plt.xlabel("Yıllar")
plt.ylabel("Buyume Oranı")
plt.title("Türkiye' nin 2001-2019 Arası Büyüme Oranları")
plt.show()
```

```python
# Bunu tek komuttada yapabiliriz.
import matplotlib.pyplot as plt
import numpy as np

yillar = np.arange(2002, 2020)
buyume = [6.4, 5.6, 9.6, 9.0, 7.1, 5.0, 0.8, -4.7, 8.5, 11.1, 4.8, 8.5, 5.2, 6.1, 3.2, 7.5, 2.8, 0.9]

plt.plot(yillar, buyume, "-o") # -: çizgi grafiği      o: nokta grafiğini temsil eder.

plt.xticks([2002, 2004, 2006, 2008, 2010, 2012, 2014, 2016, 2018])
plt.yticks([-5.0, 0.0, 5.0, 10.0], ["-%5", "%0", "%5", "%10"])
# yticks: 1.arguman; Eksen üzerindeki değerlerin hangi noktalarda olacağı
#         2.arguman; Bu noktalarda ne yazacağı !  Eğer bu arguamanı girmezsen direk sayı değerleri işlenir yazıya.
plt.xlabel("Yıllar")
plt.ylabel("Buyume Oranı")
plt.title("Türkiye' nin 2001-2019 Arası Büyüme Oranları")
plt.show()
```

```python
# !! Zaman zaman bir dağılım grafiğinde iki özelliği x ve y eksenlerinde gösterirken bir üçüncü özelliğide renk 
# ya da büyüklük olarak göstermek isteyebiliriz.
# Bu örnek için sklearn paketinde yer alan örnek bir veri kümesinden yararlanacağım.
# Bu veri kümesinde; x: Ortalama ev başına oda sayısı
#                    y: Ortalam fiyat
#                    büyüklük: şehideki suç oranı

import matplotlib.pyplot as plot
import pandas as pd
from sklearn.datasets import load_boston

boston_housing = load_boston()
boston = pd.DataFrame(boston_housing.data)
boston.columns = boston_housing.feature_names
boston["fiyat"] = boston_housing.target

plt.scatter(boston.RM, boston.fiyat, s= boston.CRIM) # s: size argumanı
plt.xlabel("Ortalama Oda Sayısı")
plt.ylabel("Fiyat")
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,10)
y = [10, 20, 30, 70, 10, 20, 40, 50, 70, 80]

# plt.plot(x, y, "r--")      Kırmızı kesikli çizgi
# plt.plot(y)                y eksenini çizdir x ekseninde 0 dan N-1 e kadar olan sayılar yazdırır. 
# plt.plot(y, "bo")          Mavi renkte yuvarlak noktalar şeklinde bir çıktı alırız.  
# plt.plot(x, y, "v")        Aşağı yönlü üçgen şeklinde noktalar
# plt.plot(x, y, "r--v")     Aşağı yönlü üçgen şeklinde noktalar kırmızı kesikli çizgi
# plt.plot(x, y, "1")         Aşağı yönlü ok
plt.plot(x, y ,"g*--")
plt.show()

# Renkler; g:green  b:blue  r:red   c:cyan  m:magenta  y:yellow   k:black   w:white
```

```python
# Hatırlatma;
# Aynı çerçeveye birden fazla grafik çizdirmenin başka bir yöntemi; plt.subplot
import matplotlib.pyplot as plt
import pandas as pd
import numpy  as np
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Create just a figure and only one subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')

# Create two subplots and unpack the output array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)


```

```python
# Matplot lib modülünde bulunan style.use() komutu ile grafiğin önceden tanımlanmış farklı
# temalarda çizdirilmesini sağlayabilirsin.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
stock = pd.read_csv("aefes-anadolu-efes.csv")
plt.scatter(stock["volume"[0:]], stock["open"[0:]], label="part1")
plt.legend(loc="upper left")
# plt.style.use("ggplot")
#plt.style.use("fivethirtyeight")
plt.style.use("dark_background")
plt.savefig("graphic.png")
plt.show()
```
