```python
import matplotlib.pyplot as plt
import numpy as np

yillar = np.arange(2002, 2020)
buyume = [6.4, 5.6, 9.6, 9.0, 7.1, 5.0, 0.8, -4.7, 8.5, 11.1, 4.8, 8.5, 5.2, 6.1, 3.2, 7.5, 2.8, 0.9]

plt.scatter(yillar, buyume)
plt.xticks([2002, 2004, 2006, 2008, 2010, 2012, 2014, 2016, 2018])
plt.yticks([-5.0, 0.0, 5.0, 10.0], ["-%5", "%0", "%5", "%10"])
# yticks: 1.arguman; Eksen Ã¼zerindeki deÄerlerin hangi noktalarda olacaÄÄ±
#         2.arguman; Bu noktalarda ne yazacaÄÄ± !  EÄer bu arguamanÄ± girmezsen direk sayÄ± deÄerleri iÅlenir yazÄ±ya.
plt.xlabel("YÄ±llar")
plt.ylabel("Buyume OranÄ±")
plt.title("TÃ¼rkiye' nin 2001-2019 ArasÄ± BÃ¼yÃ¼me OranlarÄ±")
plt.show()
```

```python
# !! Zaman zaman bir daÄÄ±lÄ±m grafiÄinde iki Ã¶zelliÄi x ve y eksenlerinde gÃ¶sterirken bir Ã¼Ã§Ã¼ncÃ¼ Ã¶zelliÄide renk 
# ya da bÃ¼yÃ¼klÃ¼k olarak gÃ¶stermek isteyebiliriz.
# Bu Ã¶rnek iÃ§in sklearn paketinde yer alan Ã¶rnek bir veri kÃ¼mesinden yararlanacaÄÄ±m.
# Bu veri kÃ¼mesinde; x: Ortalama ev baÅÄ±na oda sayÄ±sÄ±
#                    y: Ortalam fiyat
#                    bÃ¼yÃ¼klÃ¼k: Åehideki suÃ§ oranÄ±

import matplotlib.pyplot as plot
import pandas as pd
from sklearn.datasets import load_boston

boston_housing = load_boston()
boston = pd.DataFrame(boston_housing.data)
boston.columns = boston_housing.feature_names
boston["fiyat"] = boston_housing.target

plt.scatter(boston.RM, boston.fiyat, s= boston.CRIM) # s: size argumanÄ±
plt.xlabel("Ortalama Oda SayÄ±sÄ±")
plt.ylabel("Fiyat")
plt.show()

```

```python
# FarklÄ± tÃ¼rleri farklÄ± renklerde gÃ¶sterdiÄimiz bir Ã¶rneÄe bakalÄ±m.
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

iris_veri = load_iris()
iris = pd.DataFrame(iris_veri.data)
iris.columns = ["Sepal_Boy", "Sepal_En", "Petal_Boy", "Petal_En"]
iris["Cicek_Turu"] = iris_veri.target

plt.scatter(iris.Petal_Boy, iris.Petal_En, c = iris.Cicek_Turu) # c: color argumanÄ±
plt.xlabel("Petal_Boy (cm)")
plt.ylabel("Petal_En (cm)")
plt.title("Ä°ris Veri Seti")
plot.show()
```

```python
# HatÄ±rlatma;
# AynÄ± Ã§erÃ§eveye birden fazla grafik Ã§izdirmenin baÅka bir yÃ¶ntemi; plt.subplot
import matplotlib.pyplot as plt
import pandas as pd
import numpy  as np
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Create just a figure and only one subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')

# Create two subplots and unpack the output array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)


```

```python
# Matplot lib modÃ¼lÃ¼nde bulunan style.use() komutu ile grafiÄin Ã¶nceden tanÄ±mlanmÄ±Å farklÄ±
# temalarda Ã§izdirilmesini saÄlayabilirsin.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
stock = pd.read_csv("aefes-anadolu-efes.csv")
plt.scatter(stock["volume"[0:]], stock["open"[0:]], label="part1")
plt.legend(loc="upper left")
# plt.style.use("ggplot")
#plt.style.use("fivethirtyeight")
plt.style.use("dark_background")
plt.savefig("graphic.png")
plt.show()
```
