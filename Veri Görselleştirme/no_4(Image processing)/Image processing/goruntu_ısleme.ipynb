```python
import  matplotlib.pyplot as plt
harry = plt.imread("index.jpeg")
plt.imshow(harry)
plt.axis("off")
plt.show()
```

```python
harry.shape  # ilk ikisi x ve y eksenlerindeki piksel sayısı  3: renkli fotğraflardaki RGB değeridir.
```

```python
# Aynı şekli siyah beyaz çizdirelim;
# Bütün renkler kırmızı, yeşil, mavi renklerinin tonlarının karışımı ile meydana gelir.
# Siyah beyaz görüntülerde üç farklı değer yerine 0 ile 1 arasında bir değer bulunur. 0:siyah   beyaz
```

```python
# ELimizdeki görüntüyü siyah beyaz yapmak için 3 değeri tek bir değer olarak yazmalıyım
# Bunu mean yani ortalama ya da sum yani toplama ile yapabiliriz.
harry1 = plt.imread("index.jpeg")
siyah_beyaz = harry1.mean(axis=2)
plt.imshow(siyah_beyaz, cmap= "gray")
plt.axis("off")
plt.show()
```
