```python
# Python'da iki değişkenli bir fonksiyonu hesaplamak için .meshgrid() metodu kullanılır.
```

```python
# x ve y değişkenlerinden oluşan şu fonksiyona bakalım;
import numpy as np
import matplotlib.pyplot as plt
# Z = cos((sqrt(x^2) +sqrt(y^2))/π)     
# x ve y derğerleri -10 ile 10 arasında olsun.
u = np.linspace(-10, 10, 100)
v = np.linspace(-10, 10, 100)
x ,y = np.meshgrid(u, v)
Z = np.cos(np.sqrt(x**2 + y**2)/np.pi)
plt.pcolor(Z)
plt.colorbar()
# plt.imshow(Z)
plt.show()
        # Düzey Grafiği
```

```python
plt.contour(x ,y, Z)
plt.show()
```

```python
plt.contourf(x, y, Z)
plt.show()
```
