```python
# Çizgi grafik:
# Bunun için matplotlib paketinden pyplot metodunu kullanacağım.
# Türkiyenin 2001-2019 arasındaki ekonomik büyüme oranlarının grafiğini çizdirelim.
import matplotlib.pyplot as plt
import numpy as np

yillar = np.arange(2002, 2020)
buyume = [6.4, 5.6, 9.6, 9.0, 7.1, 5.0, 0.8, -4.7, 8.5, 11.1, 4.8, 8.5, 5.2, 6.1, 3.2, 7.5, 2.8, 0.9]


plt.plot(yillar, buyume) # plot(x ekseninde istediğin değişken, y ekseninde istediğin değişken)
# X ve Y eksenlerine etiket ekleyelim.
plt.xlabel("Yıllar")
plt.ylabel("Büyüme")
# Grafiğe başlık ekleyelim.
plt.title("Türkiye'nin Büyüme Oranları(2002-2019)")
plt.show()
```

```python
# Yıllar dikkatini çekmiştir. Bunu düzeltelim.
# .xticks() ve .ysticks() ile eksendeki değerleri belirleyebilirsin.
import matplotlib.pyplot as plt
import numpy as np

buyume = [6.4, 5.6, 9.6, 9.0, 7.1, 5.0, 0.8, -4.7, 8.5, 11.1, 4.8, 8.5, 5.2, 6.1, 3.2, 7.5, 2.8, 0.9]
yillar = np.arange(2002, 2020) 

plt.plot(yillar, buyume)
plt.title("Türkiye'nin Büyüme Oranları(2002-2019)")
plt.xlabel("Yıllar")
plt.ylabel("Büyüme")
plt.xticks([2002, 2004, 2006, 2008, 2010, 2012, 2014, 2016, 2018])
plt.show()
```

```python
# 2 grafiği bir arada çizdirelim;
import matplotlib.pyplot as plt
import numpy as np
y1 = np.array([25.7, 23.8, 22.4,25.6, 26.1, 27.3, 29.9, 30.1,
              31.2, 31.8, 32.2, 45.4])
y2 = np.array([22.7, 20.8, 30.4,21.6, 29.1, 30.3, 50.9, 48.1,
              12.2, 10.8, 20.2, 45.4])

yıllar = np.arange(2008, 2020)

plt.plot(yıllar, y1, "r") 
plt.plot(yıllar, y2, "b" )
plt.show()

```

```python
# Yukarıdaki 2 grafiği yan yana 2 çerçevede göstermekte mümkündür. plt.axes()
# Argümanları; x, y, genişlik, yükseklik
# x ve y argümanları grafiğin sol alt köşesinin ana grafiğe göre konumu belirtir.
# Genişlik ve yükseklik argümanları alt grafiğin ana grafiğe olan genişlik ve yükseklik oranlarıdır.
plt.axes([0.05, 0.05, 0.4, 0.9])
plt.plot(yıllar, y1, "r")
plt.axes([0.55, 0.05, 0.4, 0.9])
plt.plot(yıllar, y2, "b")
plt.show()
```

```python
# axes yöntemi ile grafiğin içinde grafik çizdirelim;
plt.axes([0.06, 0.05, 0.4, 0.9])
plt.plot(yıllar, y1, "r")
plt.axes([0.1, 0.55, 0.2, 0.2], facecolor="y")
plt.plot(yıllar, y2, "b")

plt.show()
```

```python
# Aynı çerçeveye birden fazla grafik çizdirmenin başka bir yöntemi; plt.subplot
# Argümanlar; nrows: Satır Sayısı (Oransal)  ncols: Sütun Sayısı nsubplot:Hangi alt grafiğin aktif hala getirileceği.
plt.subplot(2, 1, 1) # Aktif grafik 1
plt.plot(yıllar, y1, "r--")

plt.subplot(2, 1, 2)
plt.plot(yıllar, y1,"b")

# Grafiklerin sıkışık olmaması için;  plt.tight_layout()
plt.tight_layout()
plt.show()
```

```python
plt.plot(yıllar, y1, "r", label="HHG")
plt.plot(yıllar, y2, "b--", label="MEG")
plt.legend(loc= "upper left")
plt.show()
```

```python
# Grafiğe metin eklemek için plt.annotate xy koordinatlarını belirt .
plt.plot(yıllar, y1, "r--")
plt.plot(yıllar, y2, "b:")
plt.annotate("Text1", xy=(2014, 51))
plt.annotate("Text2" , xy=(2010,31))
plt.show
```

```python
# Kaggle dan dataset indirip grafik çizdirelim.
import numpy as np
import pandas as pd
import matplotlib.pyplot
stock = pd.read_csv("aefes-anadolu-efes.csv")
```

```python
plt.plot(stock["open"])
plt.xlabel("Years(2003-2020)")
plt.ylabel("Open Stock")

plt.xticks([])
plt.show
```

```python
stock.loc[2000:3000, "open"].plot()
plt.show()
```

```python
stock.plot()
plt.show()
```

```python
# x ve y eksenlerinin limitini belirlemek için;
# plt.axis([xmin, xmax, ymin, ymax])
# plt.xlim([xmin, xmax])   plt.ylim([ymin, ymax])
# Axis meotodu;
# axis("off") : Eksen çizgileri ve etiketleri görülmez.
# axis("equal"): Eksen uzunlukları  eşit olur.
# axis("square"): Grafik kare şeklinde çizdirilir.
# axis("tight") : x ve y ekseni limitlerini bütün verileri gösterecek şekilde düzenler.
```

```python
stock["low"].plot(color="red", style= "-")
stock["high"].plot(color="green", style=".")
plt.axis([1000, 3000, 15, 30])
plt.show()
```

```python
# Farklı bir çizim;
stock["adjclose"].plot(kind= "area", title="Alan Grafiği")
plt.show()
```

```python
stock.loc[:, ["adjclose", "open"]].plot(kind= "area", title="Alan Grafiği")
plt.show()
```
