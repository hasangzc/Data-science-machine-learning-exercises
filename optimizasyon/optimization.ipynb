```python
import numpy as np
from scipy.optimize import minimize
# Hacmi maksimize etmek istiyoruz. Neden minimize fonksiyonunu kullandık?
# Herhangi bir fonksiyonu maksimize etmek ile - lisini minimize etmek aynıdır.
# max(f(x)) <=> min(-f(x))
def calcVolume(x): 
    length = x[0]
    width = x[1]
    height = x[2]
    volume = length*width*height
    return volume


def calcSurface(x):
    length = x[0]
    width = x[1]
    height = x[2]
    surface_area = 2*length* width + 2*length*height + 2*width*height
    return surface_area


def objective(x): # Amaç fonksiyonu
    return -calcVolume(x)


def constrain1(x): # Kısıt fonksiyonu
    return 10-calcSurface(x)

# Minimize fonksiyonu kısıtları bir dictionary olarak alır. dict = {kısıt_tipi:"", function: function_name}
cons = ({"type": "ineq", "fun": constrain1})

# Optimizasyon başlangıç noktasından türev yönünde ilerleyerek en iyi çözümü bulmaya çalışır.
# Başlangıç noktalarını (Initial points) belirleyelim.
lengtGuess = 1
widthGuess = 1
heightGuess = 1

x0 = np.array([lengtGuess, widthGuess, heightGuess])

# Optimizasyon işlemine başlamadan önce problemimi iyi tanıyıp ona göre bir çözücü(solver) bulmam gerekir.
# Minimize fonksiyonu içerisinde birçok solver(algoritma) var. 
# Problemimi incelediğimde nonlinear ve kısıtlı bir problemi çözecek bir solver' a ihtiyacım  var.
# Neden nonlinear? amaç foknsiyonun: L*W*H Yani kübik bir ifade
#                  kısıt fonksiyonun: 2*L* W + 2*L*H + 2*W*H 2. dereceden bir fonksiyon
# SLSQP Solverinı kullanacağım.
sol = minimize(objective, x0, method="SLSQP", constraints=cons, options={"disp": True})

xopt = sol.x
volume_opt = -sol.fun
surface_area_opt = calcSurface(xopt)

print("length:", str(xopt[0]))
print("width:", str(xopt[1]))
print("height:", str(xopt[2]))
print("volume:", str(volume_opt))
print("Surface area:", str(surface_area_opt))
```
