```python
# 1.1) En Küçük Kareler Yöntemi
# Gözetimli öğrenme; Belirli bağımsız değişkenleri kullanarak bağımlı bir değişken tahmin etmedir.
# Bu bağımlı değişken sayısal bir değişken ise Regresyon yani Bağlanım Analizi olarak adlandırılır.
# Bağımsız değişkenler ile bağımlı değişken arasında fonksiyonel bir ilişki bulunmaya çalışılır.
```

```python
# Bağımsız değişkenimiz x ve bağımlı değişkenimiz y olsun.
# Amacımız y değişkenini x türünden açıklamak.
# Bu açıdan bulacağın modelin 3 soruya cevap vermesi gerekiyor.
# 1) 2 değişken arasında hiçbir zaman kesin bir ilişki olmadığına göre x dışındaki faktörlerin y değişkeni
# üzerindeki etkisini nasıl gözlemleriz?
# 2) x ve y arasındaki fonksiyonel ilişki nedir?
# 3) x ve y arasında ceteris paribus(yani diğer her şey sabitken) bir ilişkiyi nasıl bulabiliriz?
```

```python
# x ve y arasındaki ilişkiyi gösteren matematiksel bir denklem;
# Simle Linear Regression(Basit Doğrısal Regresyon)
# y = b0 + b1*x + u
# u: Hata Terimi ----> y değişkenini  etkileyen x dışındaki tüm faktörler.
# b0 ve b1 katsayılarını hiçbir zaman bilemeyeceğiz. Çünkü bu denklem 2 değişken arasında olduğunu varsaydığımız
# eşitliği gösterir.
# yi = (b^)0 + (b1^)xi
# Değişkenler üzerindeki şapka bunların tahmin edilen değerler olduğunu gösterir.
# yi ve xi değişkenlerinde yer alan i harfi ise gözlem sırasını belirtir.
```

```python
from sklearn import datasets
boston = datasets.load_boston()
boston.keys()
```

```python
print(boston.DESCR)
```

```python
import pandas as pd
# Bağımlı değişkenleri tanımlayalım;
boston_df = pd.DataFrame(boston.data)
boston_df.columns = boston.feature_names
# Bağımsız değişken;
boston_df["MEDV"] = boston.target
boston_df.head()
```

```python
# Şimdi konut fiyatları(MEDV) ile (LSTAT)mahallede yaşayan düşük gelirli ailelerin yüzdelik oranı ise;
# medv arttıkça lsat ın düşmesi lazım.
# Bakalım;
import matplotlib.pyplot as plt
plt.scatter(boston_df["LSTAT"], boston_df["MEDV"], alpha=0.5)
plt.xlabel("LSTAT")
plt.ylabel("MEDV")
plt.show()
```

```python
# Bağımısız değişen X(LSTAT) olsun bağımlı değişken Y(MEDV) olsun.
lstat = boston_df[["LSTAT"]]
lstat
```

```python
lstat = boston_df[["LSTAT"]].values
medv = boston_df[["MEDV"]].values
# Sklearn içindeki linear_model modülünde yer alan .LinearRegression(), fit(), predict() fonksiyonlarını kullanacağız.
# Bunlardan ilk ikisini istenen değişkenleri kullanarak doğrusal bir model oluşturmak için kullanırızç
# .predict() i ise bu modeli kullanarak tahminde bulunmak için kullanacağız.
import numpy as np
from sklearn import linear_model

model0 = linear_model.LinearRegression()
model0.fit(lstat, medv)
```

```python
# Aynı şeyi şöyle de yapabiliriz.
from sklearn.linear_model  import LinearRegression
model1 = LinearRegression()
model1.fit(lstat, medv)
# Şimdi modelimi kullanarak regresyon doğrumu çizdireceğim.
# model.predict(lstat) ifade oluşturduğum modele göre tahmini medv değerlerini verir.
import matplotlib.pyplot as plt
plt.scatter(lstat, medv, color="blue", alpha=0.7)
plt.plot(lstat, model1.predict(lstat), color="red")
plt.xlabel("LSTAT")
plt.ylabel("MEDV")
plt.show()
```

```python
# Peki bulduğum modeldeki B0 ve B1 değerlerini nasıl görebilirim. Yukarıda yer alan denklemde; 
# B0 doğrunun Y eksenini kestiği noktayı B1 de doğrunun eğimini gösterir. 
```

```python
from sklearn import datasets
boston = datasets.load_boston()
boston.keys()
import pandas as pd
# Bağımlı değişkenleri tanımlayalım;
boston_df = pd.DataFrame(boston.data)
boston_df.columns = boston.feature_names
# Bağımsız değişken;
boston_df["MEDV"] = boston.target
lstat = boston_df[["LSTAT"]].values
medv = boston_df[["MEDV"]].values
from sklearn.linear_model  import LinearRegression
model1 = LinearRegression()
model1.fit(lstat, medv)

# B0
model1.intercept_
```

```python
# B1
model1.coef_
```

```python
# Modelimiz lstat ve medv arasında şu şekilde bir ilişki hesapladı;
medv = 34.55 - 0.95*lstat
```

```python
# Şimdiye kadar tekli doğrusal bağlanım gördük. Ama doğrusal bağlanımda genellikle birden fazla bağımsız değişken 
# olur.
# Bu defa bağımsız değişkenler LSTAT VE AGE olsun. AGE değişkeni sahipleri tarafından kullanılan, 1940'tan önce
# inşa edilmiş konutların oranını gösteriyor.
# Bu durumda şu şekilde bir model bulmak istiyoruz;
# medv = (B^)0 + (B^1)*lstat + (B^2)*age
from sklearn import datasets
from sklearn.linear_model import LinearRegression
x = boston_df[["LSTAT", "AGE"]].values
y = boston_df[["MEDV"]].values
model2 = LinearRegression()
model2.fit(x, y)
model2.intercept_
```

```python
model2.coef_
# Hesaplan Model;
# medv = 33.22 - 1.03*lstat + 0.0345*age
```

```python
# Şimdide veri kümesindeki tüm bağımsız değişkenleri kullanalım.
x = boston_df.drop("MEDV", axis=1)
y = boston_df["MEDV"]
model3 = LinearRegression()
model3.fit(x, y)
model3.intercept_
```

```python
model3.coef_
```

```python
# Seaborn modülüneki bazı metodlarlada doğrusal regresyon grafiği çizdirebilirim
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

car_crashes = sns.load_dataset("car_crashes")
sns.lmplot(x="alcohol", y="total", data=car_crashes) # lmplot = linear model plot
plt.show()
# NOT! Seaborn modülü pandas ve matplotlib modülleri üzerinde kuruldu bu  yüzden bu modülü yüklemeden önce
# diğer 2 modülü yüklemeliyim.
# NOT! lmplot() fonksiyonunda kullanılacak veri kümesi DataFrame formatında olmalı.
```

```python
iris  = sns.load_dataset("iris")
sns.lmplot(x="sepal_length", y="sepal_width", hue="species", data=iris)
plt.show()
```

```python
iris  = sns.load_dataset("iris")
sns.lmplot(x="sepal_length", y="sepal_width", col="species", data=iris)
plt.show()
```

```python
iris  = sns.load_dataset("iris")
sns.lmplot(x="sepal_length", y="sepal_width", row="species", data=iris)
plt.show()
```

```python
# Seaborn ile kalan grafiği çizdirebilirim.
# Doğrusal regresyon modellerinde kalanların beklenen değeri 0'dır.
iris  = sns.load_dataset("iris")
sns.residplot(x="sepal_length", y="sepal_width", color="red", data=iris)
plt.show()
```
